```python
import cv2 
import numpy as np

img = cv2.imread('lenna.png',cv2.IMREAD_GRAYSCALE)
if img is None:
    print('Image load failed!')

bImg = cv2.blur(img,(5,5)) 

suming = cv2.integral(img)
bImg2 = np.zeros((img.shape[0],img.shape[1]))

#write filtering with 5*5 using suming
for y in range(0, img.shape[0]-4):
    for x in range(0, img.shape[1]-4):
      bImg2[y+2, x+2] = (suming[y+5,x+5] - suming[y+5,x] - suming[y,x+5] + suming[y,x]) / 25

bImg2 = (bImg2 - np.min(bImg2)) / (np.max(bImg2) - np.min(bImg2)) * 255  
bImg2 = bImg2.astype(np.uint8) 

titles = ['Original Image','Blurred','With IntegralImg']
images = [img, bImg, bImg2]

for i in range(3):
    cv2.imshow(titles[i],images[i])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#=======if u want check image, just use change _ to img name=======
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读入原图像
img_soccer = cv2.imread('soccer.jpg')

# 获取图像的红色通道
img_soccer_red = img_soccer[:,:,2]

# 对红色通道应用 Otsu's Binarization
_, img_soccer_binary = cv2.threshold(img_soccer_red, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print('Soccer image red channel - Optimal threshold:', _)

# 直方图
plt.hist(img_soccer_red.ravel(),256,[0,256])
plt.title('Histogram - Soccer Red Channel')
plt.show()

# 读取第二个图片
img_rose = cv2.imread('rose.png')

# 转换图像至灰度图
img_rose_gray = cv2.cvtColor(img_rose, cv2.COLOR_BGR2GRAY)

# 对灰度图 Ostsu's Binarization
_, img_rose_gray_binary = cv2.threshold(img_rose_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print('Rose grayscale image - Optimal threshold:', _)

# 直方图
plt.hist(img_rose_gray.ravel(),256,[0,256])
plt.title('Histogram - Rose Gray Channel')
plt.show()

# 获取图像红色通道
img_rose_red = img_rose[:,:,2]

# 对红色通道应用 Ostsu's Binarization
_, img_rose_red_binary = cv2.threshold(img_rose_red, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print('Rose image red channel - Optimal threshold:', _)

# 直方图
plt.hist(img_rose_red.ravel(),256,[0,256])
plt.title('Histogram - Rose Red Channel')
plt.show()


```
